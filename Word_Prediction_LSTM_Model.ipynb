```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
#Our task is to build a model that can predict words for us, this is essentially a word generator
# we first tokenize the data , which assigns a uniqe number to every unique word thereby tokenizing every word in the dataset
# we can use the word index function to look at the word dictionary

tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
# our main task in the first half is to create the input - output dataset from the word document
# the input data set contains the vector representation of sentence pieces so for example we have a sentence "I am Aryan"
# the input dataset will have "I" , "I am" and "I am Aryan" in the input dataset ,the output dataset contains the next word that comes in the sentence piece
# here we first split the input text based on \n so we divide it into sentences and then give the vector version of the sentences using text_to_sequence method
# we then start a loop to get sentence slices starting the first word and going until the end of the sentence
# the data set input_sequences is built as such that from index 0 - {len(sentence)-1} are the entries belonging to the input data set
# the last entry gives us the output of every sentence slice


input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences[0:20]
```

```python
max_len = max([len(x) for x in input_sequences])
#the vectors we obtain as the input sequence are not of the same length , we use padding to make them equal the size of the highest dimensional vector
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
print(padded_input_sequences)
#nested array of padded input sentence piece vectors
```

```python
output_nested_vector=[]
for vector in padded_input_sequences:
  x=vector[:-1]
  output_nested_vector.append(x)
print(output_nested_vector)
X=padded_input_sequences[:,:-1]
print(x)
```

```python
Y = padded_input_sequences[:,-1]
print(Y)
Y.shape
```

```python
X.shape
# there are 2541 sentences and each sentence is padded to be of length 185
# when we are converting the data to a categorical regression problem we need to provide the same number of channels as the number of distinct words
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(Y,num_classes=820)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
# after creating the input and output vectors we initialize the architecture of the model
# the first layer is an embedding layer, the embedding layer assignes a (100 X 1) vector to every tokenized word
# also we know that every sentence is represented by a padded 56X1 vector.
# every entry in the vector represents a tokenized word or the padding 0 now the embedding layer assignes a 100x1 vector to every tokenized vector entry
# then we send the output of the embedding layer to a 150 node LSTM network
# so in the LSTM input we send a 100 x 1 vector where each vector represents a tokenized word in the document which is in a  sentence vector of length 56 X 1
# the output of the LSTM layer is then sent through a dense layer where we use categorical cross entropy to find the output which is given by a hard-encoded vector
# so if the model predicts the next word to be of the tokenized index of 4 then the output is [0,0,0,1,0...0] we will get a 1 at that index
model = Sequential()
model.add(Embedding(820, 100, input_length=185))
model.add(LSTM(150))
model.add(Dense(820, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
model.summary()
model.fit(X,y,epochs=3)
```

```python
import time
import numpy as np
text = "I am"

for i in range(9):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  padded_token_text = pad_sequences([token_text], maxlen=185, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
